```python
from utils.Similarity import sim ## Import the sim function from utils

with open("log/Answers_llm.txt", "r", encoding="utf-8") as f:
    sentences1 = f.readlines()
with open("log/Answers_web.txt", "r", encoding="utf-8") as f:
    sentences2 = f.readlines()    
st_list, bert_list, i_list = [], [], []
for i in range(0, 100, 1):
    i = i / 100
    n_true_st, n_true_bert = sim(sentences1, sentences2, i) ## Test and get the similarity by threshold from 0 to 1 with step 0.01
    n_true_st, n_true_bert = n_true_st / 200, n_true_bert / 200
    st_list.append(n_true_st)
    bert_list.append(n_true_bert)
    i_list.append(i)
    print(str(i) + ': ' + str(n_true_st * 100) + '%, ' + str(n_true_bert * 100) + '%')
    # print("Number of Correct(SentenceTransformer): " + str(n_true_st))
    # print("Number of Correct(BERT): " + str(n_true_bert))
## Reverse the list
st_list.reverse()
bert_list.reverse()
```

```python
## Plot the graph of correctness-confidence relation
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
ax.plot(i_list, st_list, color='r', label='SentenceTransformer')
ax.plot(i_list, bert_list, color='b', label='Bert')
ax.set_xlim(0, 1)
ax.set_xticks(np.arange(0, 1.1, 0.1))
ax.set_xticklabels(['%.2f' % i for i in np.arange(1, -0.1, -0.1)])
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_xlabel('Confidence')
ax.set_ylabel('Correctness')
ax.set_title('Correctness-Confidence')
ax.legend()
plt.show()
plt.savefig('sim.png')
```
